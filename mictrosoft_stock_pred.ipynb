### ARIMA

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
from itertools import product
import warnings
warnings.filterwarnings('ignore')

# Step 1: Split the data into training and testing sets
train = new_df[new_df.index < '2025-01-01']
test = new_df[new_df.index >= '2025-01-01']

print(f"Training data: {len(train)} observations")
print(f"Test data: {len(test)} observations")

# Step 2: Check for stationarity on the training data
def check_stationarity(series, name="Series"):
    result = adfuller(series)
    print(f'\n{name} Stationarity Test:')
    print(f'ADF Statistic: {result[0]:.6f}')
    print(f'p-value: {result[1]:.6f}')
    print(f'Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value:.3f}')

    if result[1] <= 0.05:
        print("Series is stationary")
        return True
    else:
        print("Series is non-stationary")
        return False

# Check original series
is_stationary = check_stationarity(train['Close'], "Original Close Price")

# If not stationary, difference the series
if not is_stationary:
    train_diff = train['Close'].diff().dropna()
    is_diff_stationary = check_stationarity(train_diff, "First Differenced Series")
    d = 1
    data_for_acf_pacf = train_diff
else:
    train_diff = train['Close']
    d = 0
    data_for_acf_pacf = train['Close']

# Step 3: Plot ACF and PACF to determine p and q
plt.figure(figsize=(15, 6))
plt.subplot(131)
plt.plot(train['Close'][-100:])  # Plot last 100 observations
plt.title('Last 100 Training Observations')
plt.xlabel('Date')
plt.ylabel('Close Price')

plt.subplot(132)
plot_acf(data_for_acf_pacf, ax=plt.gca(), lags=40)
plt.title('ACF Plot')

plt.subplot(133)
plot_pacf(data_for_acf_pacf, ax=plt.gca(), lags=40)
plt.title('PACF Plot')
plt.tight_layout()
plt.show()

# Step 4: Grid search for best ARIMA parameters
def evaluate_arima_model(data, arima_order):
    try:
        model = ARIMA(data, order=arima_order)
        model_fit = model.fit()
        return model_fit.aic
    except:
        return float('inf')

# Define parameter ranges
p_values = range(0, 4)
d_values = [d]  # Use the determined d value
q_values = range(0, 4)

print(f"\nSearching for best ARIMA parameters...")
print(f"Testing combinations with d={d}")

best_aic = float('inf')
best_params = None
results = []

for p in p_values:
    for q in q_values:
        if p == 0 and q == 0:
            continue  # Skip (0,d,0) model

        order = (p, d, q)
        try:
            aic = evaluate_arima_model(train['Close'], order)
            results.append((order, aic))
            print(f"ARIMA{order}: AIC={aic:.2f}")

            if aic < best_aic:
                best_aic = aic
                best_params = order
        except:
            continue

print(f"\nBest ARIMA parameters: {best_params} with AIC: {best_aic:.2f}")

# Step 5: Fit the best model
model = ARIMA(train['Close'], order=best_params)
model_fit = model.fit()
print(f"\n{model_fit.summary()}")

# Check residuals
residuals = model_fit.resid
plt.figure(figsize=(15, 10))

plt.subplot(221)
plt.plot(residuals)
plt.title('Residuals')
plt.xlabel('Time')
plt.ylabel('Residuals')

plt.subplot(222)
plt.hist(residuals, bins=30)
plt.title('Residuals Distribution')
plt.xlabel('Residuals')
plt.ylabel('Frequency')

plt.subplot(223)
plot_acf(residuals, ax=plt.gca(), lags=40)
plt.title('ACF of Residuals')

plt.subplot(224)
plot_pacf(residuals, ax=plt.gca(), lags=40)
plt.title('PACF of Residuals')

plt.tight_layout()
plt.show()

# Approach 2: Rolling forecast (more realistic)
def rolling_forecast(model, train_data, test_data, window_size=None):
    """
    Perform rolling forecast where model is retrained periodically
    """
    forecasts = []
    actual_values = []

    # Use all training data initially
    current_train = train_data.copy()

    for i in range(len(test_data)):
        try:
            # Fit model on current training data
            temp_model = ARIMA(current_train, order=best_params)
            temp_fit = temp_model.fit()

            # Forecast next value
            forecast_val = temp_fit.forecast(steps=1).iloc[0]
            forecasts.append(forecast_val)

            # Get actual value using .iloc for positional indexing
            actual_val = test_data.iloc[i]
            actual_values.append(actual_val)

            # Add actual value to training data for next iteration
            new_index = test_data.index[i]
            new_data = pd.Series([actual_val], index=[new_index])
            current_train = pd.concat([current_train, new_data])

            # Optional: limit window size to most recent observations
            if window_size and len(current_train) > window_size:
                current_train = current_train.tail(window_size)

            # Print progress every 10 iterations
            if (i + 1) % 10 == 0:
                print(f"Processed {i + 1}/{len(test_data)} forecasts...")

        except Exception as e:
            print(f"Error at iteration {i}: {e}")
            # Use previous forecast or simple forecast as fallback
            if len(forecasts) > 0:
                forecasts.append(forecasts[-1])  # Use last forecast
            else:
                forecasts.append(current_train.iloc[-1])  # Use last training value
            actual_values.append(test_data.iloc[i])

    return np.array(forecasts), np.array(actual_values)

print("Performing rolling forecast (this may take a while)...")
forecast_rolling, actual_rolling = rolling_forecast(model_fit, train['Close'], test['Close'])

# Step 7: Plot results
fig, axes = plt.subplots(2, 1, figsize=(15, 12))

# Plot 2: Rolling forecast
axes[1].plot(test.index, test['Close'], label='Actual', color='blue', linewidth=2)
axes[1].plot(test.index, forecast_rolling, label='Rolling Forecast', color='green', linestyle='--', linewidth=2)
axes[1].set_title('Rolling Forecast: Actual vs Forecasted Close Prices')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Close Price')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Step 8: Calculate metrics
def calculate_metrics(actual, predicted, method_name):
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae = mean_absolute_error(actual, predicted)
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100

    print(f"\n{method_name} Metrics:")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"MAPE: {mape:.2f}%")

    return rmse, mae, mape

# Calculate metrics for both approaches
calculate_metrics(test['Close'], forecast_rolling, "Rolling Forecast")

# Step 9: Additional analysis - forecast confidence intervals
forecast_result = model_fit.get_forecast(steps=len(test))
forecast_ci = forecast_result.conf_int()

plt.figure(figsize=(15, 8))
plt.plot(train.index[-50:], train['Close'][-50:], label='Training Data (Last 50)', color='gray', alpha=0.7)
plt.plot(test.index, test['Close'], label='Actual Test Data', color='blue', linewidth=2)
plt.plot(test.index, forecast_result.predicted_mean, label='Forecast', color='red', linestyle='--', linewidth=2)
plt.fill_between(test.index,
                 forecast_ci.iloc[:, 0],
                 forecast_ci.iloc[:, 1],
                 color='red', alpha=0.2, label='Confidence Interval')
plt.title('ARIMA Forecast with Confidence Intervals')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nFinal Model: ARIMA{best_params}")
print("Note: If forecasts are still too flat, consider:")
print("1. Using external variables (ARIMAX)")
print("2. Trying machine learning models")
print("3. Ensemble methods")
print("4. The data might be close to a random walk (inherently unpredictable)")
```

### LSTM


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load and clean dataset
df = pd.read_csv("MSFT.csv", skiprows=2)
df.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
df['Date'] = pd.to_datetime(df['Date'])
df = df[['Date', 'Close']]
df = df.sort_values('Date')

# Scale close prices to [0, 1]
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df['Close'].values.reshape(-1, 1))

# Create sequences
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(time_step, len(data)):
        X.append(data[i-time_step:i])
        y.append(data[i])
    return np.array(X), np.array(y)

X, y = create_dataset(scaled_data)

# Split into training and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape input to [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint

model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

checkpoint = ModelCheckpoint(
    filepath='best_lstm_model.h5',
    monitor='val_loss',
    verbose=1,
    save_best_only=True,
    mode='min'
)

history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=64,
    validation_data=(X_test, y_test),
    callbacks=[checkpoint]
)
```

```python
y_pred = model.predict(X_test)
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(y_test, label='Actual', color='green')
plt.plot(y_pred, label='Predicted', color='orange', linestyle='--')
plt.title('LSTM Model: Actual vs Predicted on Test Data')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
